```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
```

```python
# Step 1: Define the dataset (observed sample)
data = np.array([65, 72, 68, 75, 80, 62, 71, 78, 74, 69])  # Exam scores of 10 students

# Step 2: Define the population mean (hypothesized mean)
population_mean = 70  # Null Hypothesis (Hâ‚€): The average class score is 70

# Step 3: Perform the one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, population_mean)
#$ci = stats.t.interval(0.95, df=len(data)-1, loc=np.mean(data), scale=stats.sem(data))

# Step 4: Display the results
print(f'Sample Mean: {np.mean(data):.2f}')
print(f'Hypothesized Population Mean: {population_mean:.2f}')
print(f't-Statistic: {t_statistic:.2f}')
print(f'Degrees of Freedom: {len(data) - 1}')
print(f'p-Value: {p_value:.4f}')
#print(f'Confidence Interval: [{ci[0]:.2f}, {ci[1]:.2f}]')
```

```python
# Step 5: Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f'Reject the null hypothesis. The average class score is significantly different from {population_mean:.2f}.')
else:
    print(f'Fail to reject the null hypothesis. The average class score is not significantly different from {population_mean:.2f}.')
```

```python
# Step 6: Visualize the data
plt.figure()
plt.bar(1, np.mean(data), color="skyblue", label='Sample Mean')
plt.errorbar(1, np.mean(data), yerr=np.std(data), fmt='k', linewidth=1.5)
plt.axhline(y=population_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Sample Mean vs Hypothesized Mean')
plt.xlim(0.5,1.5)
plt.xlabel('Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```

â€¢	Hypothesized mean: 5.5 <br>

o	Calculate the t-statistic and p-value.<br>
o	Interpret and visualize the result at a 0.05 significance level.<br>


```python
# One-sample t-test in Python (Colab)

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Data
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5

# Sample statistics
n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Calculate t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

print(f"Sample mean = {sample_mean:.3f}")
print(f"Sample std = {sample_std:.3f}")
print(f"t-statistic = {t_statistic:.3f}")
print(f"p-value = {p_value:.3f}")

# Interpretation at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: the mean is significantly different from 5.5")
else:
    print("Fail to reject the null hypothesis: no significant difference from 5.5")


# Create distribution of t under H0
x = np.linspace(-4, 4, 500)
y = stats.t.pdf(x, df)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="t-distribution (df=5)")

# Critical region
t_crit = stats.t.ppf(1 - alpha/2, df)
plt.axvline(t_crit, color="red", linestyle="--", label=f"Critical value Â±{t_crit:.2f}")
plt.axvline(-t_crit, color="red", linestyle="--")

# Observed t
plt.axvline(t_statistic, color="blue", linestyle="-", label=f"Observed t = {t_statistic:.2f}")

plt.title("One-sample t-test Visualization")
plt.xlabel("t-value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
